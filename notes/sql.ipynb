Relationships   
Columns in one table may relate to a column in another table.  
For example: An id column in a posts table may relate to the user id of the user who posted it in the user table. 

One-to-One:
One user has one profile. 
One person has one passport

One-to-Many:
One user has many posts. 
One author writes many books

Many-to-Many:
Many users like many posts. 
Many students take many classes

To connect 2 columns you need a primary key and a foreign key. 
Eg: passports.user_id (FK) -> users.id (PK) (One - One if both are unique)

Types of Joins.  

INNER: join and show results where 2 data properties match 
LEFT: show all results from left table and matches from right table
